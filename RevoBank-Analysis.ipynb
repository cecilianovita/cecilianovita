#DATA CLEANING

##Cleaning Dirty Data

```python
#Copy data frame
df2=df.copy()
```

```python
#Change invalid data type
df2['ACCOUNT_ID'] = df2['ACCOUNT_ID'].astype(str)
df2['BIRTH_DATE'] = pd.to_datetime(df2['BIRTH_DATE'], format = '%Y-%m-%d')
```

```python
#Remove Acct ID Duplicated Value
df2 = df2.drop_duplicates(subset = ['ACCOUNT_ID'])
```

```python
#Remove MCC invalid value
df2=df2[~(df2['MCC'] == '5311-DEPARTMENT STORES')]
```

```python
#Replace REVOSH MKTPLC with REVOSHOP
df2['MERCHANT_NAME'] = df2['MERCHANT_NAME'].replace('REVOSH MKTPLC', 'REVOSHOP')
df2['MERCHANT_NAME'].value_counts()
```

```python
#Drop row of the invalid Value in Merchant Name EL CORTE INGLES
#EL CORTE INGLES IS on the same row with dropped row before on syntax [22]
df[df['MERCHANT_NAME'] == 'EL CORTE INGLES']
```

```python
#Replace row of AVG_PROMO_TXN_AMT_L6M missing Value with 0
df2['AVG_PROMO_TXN_AMT_L6M'] = df2['AVG_PROMO_TXN_AMT_L6M'].replace(-1,0)
df2['AVG_PROMO_TXN_AMT_L6M'].min()
```

```python
#Drop MOB unusual Value
df2=df2[~(df2['MOB'] == 570)]
```

```python
##Drop row of the invalid Value in Merchant Name TOKTOKSHOP
df2=df2[~(df2['MERCHANT_NAME'] == 'TOKTOKLIVE')]
```

```python
#check on data profile
df2.info()
```

# Add collumn to help EDA Analysis in Promotion


```python
#Add col. for total sales in Promo & All Trx
df2['TOTAL_SALES_L6M'] = (df['TXN_CNT_L6M']*df['AVG_TXN_AMT_L6M']).astype(float)
df2['TOTAL_SALES_LTM'] = (df['TXN_CNT_LTM']*df['AVG_TXN_AMT_LTM']).astype(float)
df2['TOTAL_SALES_PROMO_L6M'] = (df['PROMO_TXN_CNT_L6M']*df['AVG_PROMO_TXN_AMT_L6M']).astype(float)
df2['TOTAL_SALES_PROMO_LTM'] = (df['PROMO_TXN_CNT_LTM']*df['AVG_PROMO_TXN_AMT_LTM']).astype(float)
df2.head()
```

```python
#Add collumn to measure % of Total Sales compared to Total Sales Promo within 6 month
#Value of 1 means that user id is Promo sensitive, 0 means not promo sensitive
df2['PROMO_SENSITIVE_L6M'] = (df2['TOTAL_SALES_PROMO_L6M']/df2['TOTAL_SALES_L6M']>.5).astype(int)
df2['PROMO_SENSITIVE_L6M'].value_counts()
```

```python
#Add columns to measure total promotion cost
df2['TOTAL_PROMO_COST_L6M'] = df2['TXN_CNT_L6M'] * 1.5
df2['TOTAL_PROMO_COST_LTM'] = df2['TXN_CNT_LTM'] * 1.5
```

```python
#Add col for measure total Revenue generated
df2['TOTAL_REVENUE_L6M'] = df2['TOTAL_SALES_L6M'] * .024
df2['TOTAL_REVENUE_LTM'] = df2['TOTAL_SALES_LTM'] * .024
df2['TOTAL_REVENUE_PROMO_L6M'] = df2['TOTAL_SALES_PROMO_L6M'] * .024
df2['TOTAL_REVENUE_PROMO_LTM'] = df2['TOTAL_SALES_PROMO_LTM'] * .024
```

# EDA SPENDING PATTERN & PROMO PROFITABILITY ANALYSIS



```python
#Calculate estimated Return on investment from PROMO
a = df2['TOTAL_REVENUE_L6M'].sum()
b = df2['TOTAL_REVENUE_PROMO_L6M'].sum()
c = df2['TOTAL_PROMO_COST_L6M'].sum()
d = df2['TOTAL_REVENUE_LTM'].sum()
e = df2['TOTAL_REVENUE_PROMO_LTM'].sum()
f = df2['TOTAL_PROMO_COST_LTM'].sum()



promo_roi = pd.DataFrame({
    'Metrics': ['TOTAL REVENUE', 'TOTAL REVENUE PROMO', 'PROMO COST', 'ROI'],
    'L6M': [a,b,c,b/c],
    'LTM': [d,e,f,e/f]
})

print (promo_roi)
```

```python
#Calculate Sales on promo sensitive Accounts
# PROMO SENSITIVE LEGEND : 1 = Sensitive, 0 = Non Sensitive

COUNT_SENSITIVE_ACCOUNT = df2.loc[df2['PROMO_SENSITIVE_L6M'] == 1, 'ACCOUNT_ID'].nunique()
COUNT_NONSENSITIVE_ACCOUNT = df2.loc[df2['PROMO_SENSITIVE_L6M'] == 0, 'ACCOUNT_ID'].nunique()
SENSITIVE_SALES_PROMO = df2.loc[df2['PROMO_SENSITIVE_L6M'] == 1, 'TOTAL_SALES_PROMO_L6M'].sum()
SENSITIVE_ALL_SALES = df2.loc[df2['PROMO_SENSITIVE_L6M'] == 1, 'TOTAL_SALES_L6M'].sum()
NONSENSITIVE_SALES_PROMO = df2.loc[df2['PROMO_SENSITIVE_L6M'] == 0, 'TOTAL_SALES_PROMO_L6M'].sum()
NONSENSITIVE_ALL_SALES = df2.loc[df2['PROMO_SENSITIVE_L6M'] == 0, 'TOTAL_SALES_L6M'].sum()

promo_sales = pd.DataFrame({
    'Metrics': ['Count of account','Sales Promo', 'All Sales'],
    'Sensitive': [COUNT_SENSITIVE_ACCOUNT,SENSITIVE_SALES_PROMO,SENSITIVE_ALL_SALES],
    'Non-Sensitive': [COUNT_NONSENSITIVE_ACCOUNT,NONSENSITIVE_SALES_PROMO,NONSENSITIVE_ALL_SALES],
    'All USERS' :[COUNT_SENSITIVE_ACCOUNT+COUNT_NONSENSITIVE_ACCOUNT, SENSITIVE_SALES_PROMO + NONSENSITIVE_SALES_PROMO, SENSITIVE_ALL_SALES + NONSENSITIVE_ALL_SALES]
})

print (promo_sales)
```

```python
#Create RMF Score
#r_score
r_bins = [1.999, 16.0, 17.0, 18.0, 19.0, 22.0, 28.0]
r_labels = ['1', '2', '3', '4', '5','6']
df3['r_score'] = pd.cut(df3['LAST_TXN_DAY'], bins = r_bins, labels = r_labels)

#f_score
f_bins = [0.999, 2.0, 3.0, 4.0, 5.0, 7.0, 18.0]
f_labels = ['1', '2', '3', '4', '5','6']
df3['f_score'] = pd.cut(df3['TXN_CNT_L6M'], bins = f_bins, labels = f_labels)

#m_score
m_bins = [9.999, 250.0, 330.0, 426.8, 550.0, 760.2, 8500.0]
m_labels = ['1', '2', '3', '4', '5','6']
df3['m_score'] = pd.cut(df3['TOTAL_SALES_L6M'], bins = m_bins, labels = m_labels)

df3
```

```python
#Create Low Med High Segment

df3['r_score'] = df3['r_score'].astype(float)
df3['f_score'] = df3['f_score'].astype(float)
df3['m_score'] = df3['m_score'].astype(float)

df3['r_segment'] = pd.cut(df3['r_score'], bins=[0, 3, 6], labels=['Low', 'High'], include_lowest=True)
df3['f_segment'] = pd.cut(df3['f_score'], bins=[0, 3, 6], labels=['Low', 'High'], include_lowest=True)
df3['m_segment'] = pd.cut(df3['m_score'], bins=[0, 3, 6], labels=['Low', 'High'], include_lowest=True)
df3['rfm_segment'] = df3['r_segment'].astype(str) + " " + df3['f_segment'].astype(str) + " " + df3['m_segment'].astype(str)
df3['rfm_segment'].value_counts()
```

```python

print(df3['r_segment'].value_counts())
print(df3['f_segment'].value_counts())
print(df3['m_segment'].value_counts())

```

Segmentation:


- Low R, Low F, Low M : User non priority
- Low R, High F, High M : Buy alot in the past with high amount
- High R, Low F, Low M : Recently purchase only most likely due to offered discount or special sku
- High R, High F, High M : Big Spender
- Low R, Low F, High M : One time buyer
- High R, Low F, High M : Buy in bulk probably
- Low R, High F, Low M : Buy alot in the past but low amount
- High R, High F, Low M : Buy alot of things with cheaper options




```python
#Rename Value of segmentation

df3['rfm_segment'] = df3['rfm_segment'].replace('Low Low Low', 'Infrequent Customer')
df3['rfm_segment'] = df3['rfm_segment'].replace('Low High High', 'Past Big Spender')
df3['rfm_segment'] = df3['rfm_segment'].replace('High Low Low', 'Discount Shopper')
df3['rfm_segment'] = df3['rfm_segment'].replace('High High High', 'Big Spender')
df3['rfm_segment'] = df3['rfm_segment'].replace('Low Low High', 'One Time Shopper')
df3['rfm_segment'] = df3['rfm_segment'].replace('High Low High', 'Bulk Buyer')
df3['rfm_segment'] = df3['rfm_segment'].replace('Low High Low', 'Past Low Shopper')
df3['rfm_segment'] = df3['rfm_segment'].replace('High High Low', 'Bargain Hunters')

df3['rfm_segment'].value_counts()
```
