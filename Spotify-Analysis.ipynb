##2.Compare average songs produced per artist in Indonesian Pop & Indie Music in other countries

```python
songs_count = df.groupby('Genre New')['Title'].nunique()
songs_count
```

```python
songs_count = df.groupby(['Genre New','Artist'])['Title'].nunique().groupby('Genre New').mean()
songs_count
```

```python
songs_count = df.groupby(['Genre New','Artist'])['Title'].nunique().groupby(['Genre New','Artist']).mean()
songs_count
```

```python
songs_count_df = songs_count.reset_index()
songs_count_df
```

```python
from scipy.stats import ttest_ind

indonesian_pop = songs_count_df[songs_count_df['Genre New'] == 'Indonesian Pop']['Title']
indonesian_indie = songs_count_df[songs_count_df['Genre New'] == 'Indonesian Indie']['Title']

t_stat, p_value = ttest_ind(indonesian_pop, indonesian_indie, equal_var=False)

alpha = 0.05
if p_value < alpha:
    print(f"The p-value ({p_value:.4f}) is less than alpha ({alpha}), so there is a significant difference.")
else:
    print(f"The p-value ({p_value:.4f}) is greater than alpha ({alpha}), so there is no significant difference.")
```

##3.Do statistical analysis to measure each factors impact on popularity

```python
df.info()
```

```python
#Define Variable
dependent_variable = df['Popularity']
independent_variables = df[['danceability', 'energy', 'key','loudness','mode','speechiness','acoustics','instrumentalness','liveliness','valence','tempo','duration_ms','Days_since_release','Released_after_2017','Explicit_false','Explicit_true','album','compilation','single','bing_norm_negative','bing_norm_neutral','bing_norm_positive','Top50_dummy','Top10_dummy']]

```

```python
from scipy.stats import pearsonr

#Calculate the correlation
correlation_matrix = np.corrcoef(dependent_variable, independent_variables, rowvar=False)

corr_coefficients = correlation_matrix[0, 1:]

correlation_df = pd.DataFrame({
    'Variable': independent_variables.columns,
    'Correlation': corr_coefficients
})

#Create correlation stregth
conditions = [
    (correlation_df['Correlation'].abs() >= 0.80),
    (correlation_df['Correlation'].abs() >= 0.60),
    (correlation_df['Correlation'].abs() >= 0.40),
    (correlation_df['Correlation'].abs() >= 0.20),
    (correlation_df['Correlation'].abs() >= 0.00)
]

strength_labels = ['Very Strong', 'Strong', 'Moderate', 'Weak', 'Very Weak']

correlation_df['Correlation_Strength'] = np.select(conditions, strength_labels, default='No Correlation')

#P-values column
correlation_df['P-value'] = correlation_df['Variable'].apply(lambda x: pearsonr(dependent_variable, independent_variables[x])[1])

#New column p-value significance
correlation_df['P-Value_Significant'] = np.where(correlation_df['P-value'] < 0.05, 'Yes', 'No')

correlation_df
```

```python
significant_variable = correlation_df[correlation_df['P-Value_Significant'] == 'Yes']
significant_variable
```
